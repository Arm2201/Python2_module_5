## Syntax Error, Exception Error

Syntax error is mistake with grammar in python
Exception error happens while the program is running — the syntax is fine, but something goes wrong during execution.

Difference:
Syntax error → problem in code structure (caught before running).
Exception error → problem during execution (caught after running).

Exception errors are the ones related to runtime errors.

A runtime error is an error that happens while the program is running. It means the code’s syntax is correct, but something goes wrong during execution that stops the program.

```python
# This work
def a():
    b()

def b():
    return None
a()

# This won't
def a():
    b()

a()

def b():
    return None
```

Flow of the exception The try statement works as follows.

First, the try clause (the statement(s) between the try and except keywords) is executed.
If no exception occurs, the except clause is skipped and execution of the try statement is finished.
If an exception occurs during execution of the try clause, the rest of the clause is skipped. Then if its type matches the exception named after the except keyword, the except clause is executed, and then execution continues after the try statement.
If an exception occurs which does not match the exception named in the except clause, it is passed on to outer try statements; if no handler is found, it is an unhandled exception and execution stops with a message as shown above.

Write function get_country_name(phone_number).

phone_number should be in the format of '+XXOOOOOOOOOO' where +XX is the country code and other ten O characters is the number itself. This function should return string values:

"Spain" if it begins with "+34"
Your native country if it begins with your_country_code
Optinal: contries form https://en.wikipedia.org/wiki/List_of_country_calling_codes if you are determined enough
Raise:

LookupError if the number doesn't contain any country code.
ValueError if the number contains country code, but the amount of characters are incorrect
TypeError if the telephone number is not a string
Or other exception that might happen
https://docs.python.org/3/library/exceptions.html - List of Python build-in exception types

```python
def get_country_name(phone_number: str) -> str:
    if not isinstance(phone_number, str):
        raise TypeError("Phone number must be a string type.")
    if phone_number[0] != '+':
        raise ValueError("Phone number must start with a '+' sign.")
    country_code_number = {'+1': 'USA', 
                        '+44': 'UK',
                        '+91': 'India',
                        '+7': 'Russia',
                        '+49': 'Germany',
                        '+221': 'Senegal'
                        }
    if phone_number[0:2] in country_code_number.keys() and len(phone_number) == 12:
        return country_code_number[phone_number]
    elif phone_number[0:3] in country_code_number.keys() and len(phone_number) == 13:
        return country_code_number[phone_number]
    elif phone_number[0:4] in country_code_number.keys() and len(phone_number) == 14:
        return country_code_number[phone_number]
    else:
        if 12 <= len(phone_number) <= 14:
            raise LookupError("Country code not found or invalid phone number length.")
        else:
            raise ValueError("Invalid phone number length.")
```
