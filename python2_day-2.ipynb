## Syntax Error, Exception Error

Syntax error is mistake with grammar in python
Exception error happens while the program is running — the syntax is fine, but something goes wrong during execution.

Difference:
Syntax error → problem in code structure (caught before running).
Exception error → problem during execution (caught after running).

Exception errors are the ones related to runtime errors.

A runtime error is an error that happens while the program is running. It means the code’s syntax is correct, but something goes wrong during execution that stops the program.

## Handling Exception

```python
# try and except
try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print("Result:", result)

except ValueError:
    print("Please enter a valid number.")

except ZeroDivisionError:
    print("You can’t divide by zero!")
```

```python
def divide(x, y):
    try:
        result = x // y
    except ZeroDivisionError:
        return "Cannot divide by zero."
    else:
        return result
    finally:
        print('Program ended.')
    
divide(3, 2)
divide(3, 0)
```

```python
number = int(input("Enter an even number: "))

if number % 2 != 0:
    raise ValueError(f'You need to enter an even number, {number} is not even.')
```

## Assert

```python
x = 5
assert x = 10
```

```python
def print_phone_num(phone_number):
    assert type(phone_number) == str, "Phone number should be a string."
    print(f"Phone number is: {phone_number}")
```

```python
address = input("Enter your addrerss: ")
postal_code = address.split()[-1] # assuming postal code is the last part of the address

def print_postal(code):
    assert type(code) == str, f'Postal code should be a string but get {type(code)}.'
    assert len(code) == 5, f'Postal code should be 5 characters long.'
    print(code)

print_postal(postal_code)
```
